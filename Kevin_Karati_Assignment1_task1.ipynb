```python
import pandas as pd

#Data frame from dictionary of lists
all_data_dic = {'f_name': ["David","Sarah","Beth","Kim","Suzan","Ibrahim","Max",'David','Sachin','Bing'],
                'l_name': ["Smith","Cornell","Carnaige","Xue","Smith","Hussain","Campbell",'Peter','Singh','Lee'],
                'as1_marks': [37,40,19,12,39,22,34,10,27,32],
                'as2_marks': [21,56,34,52,44,26,60,20,30,45],
                'year_enrolled': [2020,2020,2019,2020,2018,2022,2021,2018,2019,2022],
                'email': ["ds@yahoo.com","SC@hotmail.com","Bc@gmail.com","kx@yahoo.com","sS@yahoo.com",
                         "IH@hotmail.com","mc@hotmail.com","DP@gmail.com","SS@gmail.com","BL@hotmail.com"],
                'campus': ["Auckland","Auckland","Wellington","Nelson","Christchurch",
                          "Wellington","Auckland","Nelson","Christchurch","Napier"],
                'mode_delivery': [True,False,False,False,True,True,False,False,False,False],
                'additional_courses': [2,3,1,2,0,0,2,2,1,2],
                'software_skill': ["python,c++,html","c++,java,vb,css,bash","go,python","matlab,c#","html",
                                  "c#,ptython,go,matlab,sql,java,javascript","Java Script","Python,HTML,others,vb",
                                  "java,c,assembly,bash,html,css,go,Others","javaScript,matlab,c,css"]}

df = pd.DataFrame(all_data_dic)
```

```python
# Task A1. Calculate the average marks for assignments 1 and 2 for all the students.
df['average'] = (df['as1_marks'] + df['as2_marks'])/2
print("1. Average marks for assignments 1 and 2:")
print(df['average'])

# Calculate the final/total marks obtained by the students which is calculated as assignment 1 + assignment 2
df['total_marks'] = df['as1_marks'] + df['as2_marks']
print("\nTotal marks obtained by the students:")
print(df['total_marks'])
```

```python
# Task A2. Extract the first and last names of those students who are not registered in any other additional courses
df_no_course = df[df['additional_courses'] == 0]
df_no_course_names = df_no_course[['f_name', 'l_name']]
print("\n2. Names of students not registered in any other additional courses:")
print(df_no_course_names)
```

```python
# Task A3. create a new dataframe containing only the students with 4 or more software skills
software_expert_df = df[df['software_skill'] >= str(4)]

# extract the names and campuses of those students
names_and_campuses = software_expert_df[['f_name', 'l_name', 'campus']]
print("\n3. students with 4 or more software skills")
print(names_and_campuses)
```

```python
# Task A4. find the year with the highest enrolments
year_enrolments = df.groupby(['year_enrolled'])['f_name'].count()
max_year_enrolments = year_enrolments.idxmax()
print(f"\n4. The year with the highest enrolments is {max_year_enrolments}")
```

```python
# Task A5 replace the values under mode_delivery
df['mode_delivery'] = df['mode_delivery'].replace({True: 'online', False: 'onsite'})
print(df)
```

```python
# Task A6. Filter the dataframe to only include Max Campbell's row
max_campbell = df[(df['f_name'] == 'Max') & (df['l_name'] == 'Campbell')]

# Check if the conditions are met
if (max_campbell['as1_marks'] + max_campbell['as2_marks'] >= 70).bool() \
        and max_campbell['campus'].isin(['Wellington', 'Auckland']).bool() \
        and max_campbell['additional_courses'].any() > 0:
    print("Max Campbell is eligible for the summer internship program.")
else:
    print("Max Campbell is not eligible for the summer internship program.")
```

```python
# Task B1. Add a new column with total marks
df['total_marks'] = df['as1_marks'] + df['as2_marks']

# Add a new column with total grade based on total marks
df['total_grade'] = df['total_marks'].apply(lambda x: 'F' if x < 50 else
                                               'C' if x < 60 else
                                               'B-' if x < 70 else
                                               'B+' if x < 80 else
                                               'A' if x < 90 else
                                               'A-' if x < 95 else
                                               'A+')
```

```python
# Task B2 Rename the 'email' column to 'personal_email'
df = df.rename(columns={'email': 'personal_email'})

# Create the 'yoobee_email' column
df['yoobee_email'] = df.apply(lambda row: f"{row['f_name'].lower()}.{row['l_name'].lower()}@yobeecolleges.com", axis=1)

# Print the updated DataFrame
print(df)
```
