```python
import getpass

def rps_game():
    print("\nGame 1: Rock, Paper, Scissors")

    # Get player 1's choice
    print("Player 1's turn:")
    p1_choice = getpass.getpass("Enter your choice (1 = rock, 2 = paper, 3 = scissors): ")
    while p1_choice not in ["1", "2", "3"]:
        p1_choice = getpass.getpass("Invalid choice. Enter your choice (1 = rock, 2 = paper, 3 = scissors): ")
    print("Player 1 has chosen.")

    # Get player 2's choice
    print("Player 2's turn:")
    p2_choice = getpass.getpass("Enter your choice (1 = rock, 2 = paper, 3 = scissors): ")
    while p2_choice not in ["1", "2", "3"]:
        p2_choice = getpass.getpass("Invalid choice. Enter your choice (1 = rock, 2 = paper, 3 = scissors): ")
    print("Player 2 has chosen.")

    # Display choices as emojis
    emojis = {"1": "ðŸª¨", "2": "ðŸ“„", "3": "âœ‚ï¸"}
    print(f"Player 1: {emojis[p1_choice]}")
    print(f"Player 2: {emojis[p2_choice]}")

    # Determine the winner
    if p1_choice == p2_choice:
        print("It's a tie!")
        return None
    elif (p1_choice == "1" and p2_choice == "3") or \
         (p1_choice == "2" and p2_choice == "1") or \
         (p1_choice == "3" and p2_choice == "2"):
        print("Player 1 wins!")
        return 1
    else:
        print("Player 2 wins!")
        return 2

import random

def snakes_and_ladders():
    print("\nGame 2: Snakes and Ladders")

    # Initialise the board and players
    board = [" "] * 40
    player1_pos = 0
    player2_pos = 0

    # Define the snakes and ladders
    snakes = {16: 6, 20: 8, 30: 11, 39: 3}
    ladders = {5: 18, 12: 23, 15: 32, 25: 36}

    # Main game loop
    while True:

        # Player 1's turn
        input("Press Enter to roll the dice for Player 1...")
        roll = random.randint(1, 6)
        print(f"Player 1 rolled a {roll}!")
        player1_pos += roll
        if player1_pos > 39:
            print("Player 1 has reached the end of the board and wins!")
            return 1
        if player1_pos in snakes:
            print("Player 1 landed on a snake! Going down...")
            player1_pos = snakes[player1_pos]
        elif player1_pos in ladders:
            print("Player 1 landed on a ladder! Going up...")
            player1_pos = ladders[player1_pos]
        board[player1_pos] = "1"

        # Print the updated board
        print("\n" + "-" * 32)
        for i in range(4):
            row = board[i*10:i*10+10]
            print("| " + " | ".join(row) + " |")
        print("-" * 32)

        # Player 2's turn
        input("Press Enter to roll the dice for Player 2...")
        roll = random.randint(1, 6)
        print(f"Player 2 rolled a {roll}!")
        player2_pos += roll
        if player2_pos > 39:
            print("Player 2 has reached the end of the board and wins!")
            return 2
        if player2_pos in snakes:
            print("Player 2 landed on a snake! Going down...")
            player2_pos = snakes[player2_pos]
        elif player2_pos in ladders:
            print("Player 2 landed on a ladder! Going up...")
            player2_pos = ladders[player2_pos]
        board[player2_pos] = "2"
        
      # Print the updated board
        print("\n" + "-" * 32)
        for i in range(4):
            row = board[i*10:i*10+10]
            print("| " + " | ".join(row) + " |")
        print("-" * 32)
        # Clear the board
        board[player1_pos] = " "
        board[player2_pos] = " "



def number_guessing_game():
    print("Number Guessing Game")
    player1_score = 0
    player2_score = 0

    for i in range(2):
        if i == 0:
            print("player 1's turn:")
        else:
            print("player 2's turn:")
        secret_number = random.randint(1, 100)
        for j in range(5):
            guess = int(input("Guess a number between 1 and 100: "))
            if guess < secret_number:
                print("Your guess is too low.")
            elif guess > secret_number:
                print("Your guess is too high.")
            else:
                print("Congratulations! You guessed the number.")
                if i == 0:
                    player1_score += 1
                else:
                    player2_score += 1
                break
        else:
            print("Sorry, you have run out of turns. The secret number was", secret_number)
    print("Game Over")
    print("player 1 scored: ", player1_score)
    print("player 2 scored: ", player2_score)
    if player1_score > player2_score:
        print("player 1 wins!")
    elif player2_score > player1_score:
        print("player 2 wins!")
    else:
        print("The game is tied!")


# Main program loop
scores = {1: 0, 2: 0}

print("Welcome to the Board Game Olympics!")

# Play game 1
winner = rps_game()
if winner:
    scores[winner] += 1

# Play game 2
winner = snakes_and_ladders()
if winner:
    scores[winner] += 1

# Play game 3
winner = number_guessing_game()
if winner:
    scores[winner] += 1

# Determine the ultimate winner
if scores[1] > scores[2]:
    print(f"Player 1 is the ultimate winner with {scores[1]} out of 3 games won.")
elif scores[2] > scores[1]:
    print(f"Player 2 is the ultimate winner with {scores[2]} out of 3 games won.")
else:
    print("It's a tie! Both players won 1 game each.")
```
