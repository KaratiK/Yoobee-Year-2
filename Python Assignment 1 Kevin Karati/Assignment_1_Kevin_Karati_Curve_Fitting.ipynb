```python
import numpy as np
import matplotlib.pyplot as plt

# Load the data
X = np.array([107.608, 108.632, 109.773, 110.929, 112.075, 113.27, 115.094, 116.219, 117.388, 118.734, 120.445, 121.95, 123.366, 125.368, 127.852, 130.081])
Y = np.array([60.323, 61.122, 60.171, 61.187, 63.221, 63.639, 64.989, 63.761, 66.019, 67.857, 68.169, 66.513, 68.655, 69.564, 69.331, 70.551])
```

```python
# Define the functions
def f1(x):
    return 0.49*x + 8.38

def f2(x):
    return 3.254*x - 0.0117*x**2 - 155.028

def f3(x):
    return -3052.17*x + 46.702*x**2 - 0.3532*x**3 + 0.001318*x**4 - 0.00000193581*x**5 + 79040.97

def f4(x):
    return 0.8831*np.sin(1.4862-x) + 0.00206*x**2 + 36.904

# Plot the functions with the data
plt.figure(figsize=(10, 6))
plt.scatter(X, Y, label='Data points')
plt.plot(X, f1(X), label='Function 1')
plt.plot(X, f2(X), label='Function 2')
plt.plot(X, f3(X), label='Function 3')
plt.plot(X, f4(X), label='Function 4')
plt.legend()
plt.show()
```

```python
# Calculate the RMSE for each function
def rmse(predictions, targets):
    return np.sqrt(((predictions - targets) ** 2).mean())

f1_rmse = rmse(f1(X), Y)
f2_rmse = rmse(f2(X), Y)
f3_rmse = rmse(f3(X), Y)
f4_rmse = rmse(f4(X), Y)

print('Function 1 RMSE:', f1_rmse)
print('Function 2 RMSE:', f2_rmse)
print('Function 3 RMSE:', f3_rmse)
print('Function 4 RMSE:', f4_rmse)
```

```python
# Create a bar plot of the RMSE values
plt.figure(figsize=(6, 4))
plt.bar([1, 2, 3, 4], [f1_rmse, f2_rmse, f3_rmse, f4_rmse])
plt.xticks([1, 2, 3, 4], ['Function 1', 'Function 2', 'Function 3', 'Function 4'])
plt.xlabel('Function')
plt.ylabel('RMSE')
plt.title('RMSE for each function')
plt.show()
#Plot 4 is the best fit for the data
```
